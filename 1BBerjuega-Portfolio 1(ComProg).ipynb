As computers can evaluate complex formulas or repetitive calculations much faster and more accurately than a human can do manually or with a calculator, coding mathematical expressions has been utilized. In this portfolio, a personalized code with a simple example is presented to demonstrate the use of expressions. 

* Example: An expression for parabola;  $(x+5)^2=-1/3(y-2)$
    * In this example, we will try to find the geometric components of a parabola with the expression given above.  

1. Getting the vertex

```python
#getting the vertex of the parabola
#Since vertex is basically the opposite sign of what's in the parenthesis, we can do it like this: 
h= -(+5)
k= -(-2)

vertex= (h,k)

#output
print(vertex)
```

2. Getting the focal length

```python
#getting the focal length(p) using 4p=-1/3
p= (-1/3)/4    
print(p)
```

3. Locating the focus

```python
#locating the focus using (h, k=p)

p= (-1/3)/4
h=-(+5)
k= -(-2)

focus= (h, k+p)

print(focus)
```

4. Finding the directrix

```python
#using k-p to get the directrix

k= -(-2)
p= (-1/3)/4

directrix= k-p

print(directrix)
```

5. Getting the latus rectum

```python
#we will use 4*p

p= (-1/3)/4

latus_rectum = abs(4*p)

print(latus_rectum)
```

6. Geting the axis of symmetry

```python
#Since it is a vertical parabola, x=h

h= -(+5)
x=h

print(x)
```
